# PyCity Schools Analysis

* On average, students at charter schools outperformed those at district schools. This is especially true for math. Charter school studentâ€™s average math score average was 6.57 higher compared to average math score of district school students. Additionally, the average reading score of charter students was higher by 2.93. The percent of students with passing grades was also higher for charter schools - 27.07% in math and 15.79% in reading. 

* There was a very small performance difference between schools of small and medium size. However, students in large schools performed significantly lower than both small and medium schools. This is true for both math and reading. Only 69.96% of students in large schools received a passing score in math as compared to 93.55% and 93.60% of students in small and medium school students respectively. For reading, 82.77% from large school students received a passing score in reading compared to 96.10% and 96.79% of students from small and medium schools.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
numSchools=school_data["school_name"].count()
#Calculate the total number of students
total_students=school_data["size"].sum()
#Calculate the total budget
total_budget=school_data["budget"].sum()
#Calculate the average math score 
avg_math=student_data["math_score"].mean()
#Calculate the average reading score
avg_read=student_data["reading_score"].mean()
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
avg_score=(avg_math+avg_read)/2
#Calculate the percentage of students with a passing math score (70 or greater)
mathpass=student_data.loc[(student_data["math_score"]>=70),:]
num_mathpass=mathpass["grade"].count()
mathpassP=(num_mathpass/total_students)*100
#Calculate the percentage of students with a passing reading score (70 or greater)
readpass=student_data.loc[(student_data["reading_score"]>=70),:]
num_readpass=readpass["grade"].count()
readpassP=(num_readpass/total_students)*100

#Create a dataframe to hold the results
district_summary=pd.DataFrame([
    {"Total Schools":numSchools,
     "Total Students":total_students,
     "Total Budget":total_budget, 
     "Average Math Score":avg_math,
    "Average Reading Score":avg_read,
     "% Passing Math":mathpassP,
     "% Passing Reading":readpassP,
     "% Overall Passing Rate":avg_score}
])

#Apply formatting 
district_summary["Total Students"]=district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"]=district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"]=district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"]=district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["% Passing Math"]=district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"]=district_summary["% Passing Reading"].map("{:.2f}%".format)
district_summary["% Overall Passing Rate"]=district_summary["% Overall Passing Rate"].map("{:.2f}%".format)

district_summary=district_summary[["Total Schools","Total Students","Total Budget","Average Math Score","Average Reading Score",
                                 "% Passing Math","% Passing Reading","% Overall Passing Rate"]]
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Group Student Data by school and calculate scores
group_school=student_data.groupby(["school_name"])
stu_num=group_school["gender"].count()
school_math=group_school["math_score"].mean()
school_reading=group_school["reading_score"].mean()
school_mathP=student_data[student_data["math_score"]>=70].groupby(["school_name"])
school_mathPass=100*school_mathP["gender"].count()/stu_num
school_readingP=student_data[student_data["reading_score"]>=70].groupby(["school_name"])
school_readingPass=100*school_readingP["gender"].count()/stu_num
school_overallP=(school_mathPass+school_readingPass)/2

#Store score by school information in a dataframe
score_summary=pd.DataFrame({
    "Average Math Score":school_math,
    "Average Reading Score":school_reading,
    "% Passing Math": school_mathPass,
    "% Passing Reading": school_readingPass,
    "% Overall Passing Rate":school_overallP
})

#Get school information
schooldf=school_data[["school_name","type","size","budget"]]
schooldf=schooldf.rename(columns={"type":"School Type","size":"Total Students","budget":"Total Budget"})
schooldf["Per Student Budget"]=schooldf["Total Budget"]/schooldf["Total Students"]

#Merge data frames
school_summary=pd.merge(schooldf, score_summary, on="school_name", how="left")
school_summary=school_summary.set_index("school_name")
school_summary.sort_values("school_name")

#Format table
school_summaryFormatted=school_summary.copy(deep=True)
school_summaryFormatted["Total Students"]=school_summaryFormatted["Total Students"].map("{:,}".format)
school_summaryFormatted["Total Budget"]=school_summaryFormatted["Total Budget"].map("${:,.2f}".format)
school_summaryFormatted["Per Student Budget"]=school_summaryFormatted["Per Student Budget"].map("${:,.2f}".format)
school_summaryFormatted["Average Math Score"]=school_summaryFormatted["Average Math Score"].map("{:.2f}".format)
school_summaryFormatted["Average Reading Score"]=school_summaryFormatted["Average Reading Score"].map("{:.2f}".format)
school_summaryFormatted["% Passing Math"]=school_summaryFormatted["% Passing Math"].map("{:.2f}%".format)
school_summaryFormatted["% Passing Reading"]=school_summaryFormatted["% Passing Reading"].map("{:.2f}%".format)
school_summaryFormatted["% Overall Passing Rate"]=school_summaryFormatted["% Overall Passing Rate"].map("{:.2f}%".format)

school_summaryFormatted
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#Sort and display the top five schools in overall passing rate
top_schools=school_summary.sort_values("% Overall Passing Rate", ascending=False).head()
top_schoolsFormatted=top_schools.copy(deep=True)
top_schoolsFormatted["Total Students"]=top_schoolsFormatted["Total Students"].map("{:,}".format)
top_schoolsFormatted["Total Budget"]=top_schoolsFormatted["Total Budget"].map("${:,.2f}".format)
top_schoolsFormatted["Per Student Budget"]=top_schoolsFormatted["Per Student Budget"].map("${:,.2f}".format)
top_schoolsFormatted["Average Math Score"]=top_schoolsFormatted["Average Math Score"].map("{:.2f}".format)
top_schoolsFormatted["Average Reading Score"]=top_schoolsFormatted["Average Reading Score"].map("{:.2f}".format)
top_schoolsFormatted["% Passing Math"]=top_schoolsFormatted["% Passing Math"].map("{:.2f}%".format)
top_schoolsFormatted["% Passing Reading"]=top_schoolsFormatted["% Passing Reading"].map("{:.2f}%".format)
top_schoolsFormatted["% Overall Passing Rate"]=top_schoolsFormatted["% Overall Passing Rate"].map("{:.2f}%".format)
top_schoolsFormatted
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Create a pandas series for each grade. Hint: use a conditional statement.Group each series by school
df9=student_data[student_data["grade"]=="9th"].groupby(["school_name"])
nine=df9["math_score"].mean()
df10=student_data[student_data["grade"]=="10th"].groupby(["school_name"])
ten=df10["math_score"].mean()
df11=student_data[student_data["grade"]=="11th"].groupby(["school_name"])
eleven=df11["math_score"].mean()
df12=student_data[student_data["grade"]=="12th"].groupby(["school_name"])
twelve=df12["math_score"].mean()

#Combine the series into a dataframe
math_bygrade=pd.DataFrame({"9th": nine,
                           "10th": ten,
                           "11th":eleven,
                           "12th":twelve})

#Optional: give the displayed data cleaner formatting
math_bygradeFormatted=math_bygrade.copy(deep=True)
math_bygradeFormatted["9th"]=math_bygradeFormatted["9th"].map("{:.2f}".format)
math_bygradeFormatted["10th"]=math_bygradeFormatted["10th"].map("{:.2f}".format)
math_bygradeFormatted["11th"]=math_bygradeFormatted["11th"].map("{:.2f}".format)
math_bygradeFormatted["12th"]=math_bygradeFormatted["12th"].map("{:.2f}".format)
math_bygradeFormatted
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
schoolspend=school_summary[["Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing Rate"]]
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
schoolspend["Per Student Budget"]=pd.cut(schoolspend["Per Student Budget"], spending_bins, labels=group_names)
schoolspend=schoolspend.rename(columns={"Per Student Budget":"Spending Ranges (Per Student)"})
schoolspend=schoolspend.groupby("Spending Ranges (Per Student)")
scores_by_schoolspend=schoolspend.mean()
#format
scores_by_schoolspend["Average Math Score"]=scores_by_schoolspend["Average Math Score"].map("{:.2f}".format)
scores_by_schoolspend["Average Reading Score"]=scores_by_schoolspend["Average Reading Score"].map("{:.2f}".format)
scores_by_schoolspend["% Passing Math"]=scores_by_schoolspend["% Passing Math"].map("{:.2f}%".format)
scores_by_schoolspend["% Passing Reading"]=scores_by_schoolspend["% Passing Reading"].map("{:.2f}%".format)
scores_by_schoolspend["% Overall Passing Rate"]=scores_by_schoolspend["% Overall Passing Rate"].map("{:.2f}%".format)
scores_by_schoolspend
```
